*Some of the popular languages that Data Scientists use are:*

1. Python: Python is one of the most widely used programming languages in the field of data science. It offers a vast ecosystem of libraries and tools, such as NumPy, Pandas, and scikit-learn, making it ideal for data analysis, machine learning, and data visualization.

2. SQL (Structured Query Language): SQL is essential for managing and querying relational databases, which are commonly used in data science. Data scientists often use SQL to extract and manipulate data from databases to perform analytics.

3. Java: Java is not as common as Python or R in data science, but it is still used in certain applications, especially when dealing with large-scale data processing, distributed computing, and building data-related applications.


*Some of the commonly used libraries by Data Scientists include:*

- NumPy: Serving as a foundational library for numerical computations within Python, NumPy delivers essential support for handling large, multi-dimensional arrays and matrices. Additionally, it supplies a suite of mathematical functions tailored for performing operations on these arrays.

- Pandas: Pandas stands out as a robust data manipulation library, offering user-friendly data structures like DataFrames and Series. This simplifies data analysis and manipulation tasks in Python, making them more approachable for users.

- Matplotlib: Matplotlib enjoys widespread adoption as a Python library for generating static, animated, or interactive visualizations in both 2D and 3D formats. Its significance is particularly pronounced in the realm of data visualization and presentation.

- Seaborn: Seaborn represents a data visualization library that builds upon the foundation of Matplotlib. It provides a high-level interface geared towards crafting informative and aesthetically appealing statistical graphics.
